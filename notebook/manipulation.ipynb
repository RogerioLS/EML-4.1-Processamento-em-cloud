```python
import requests

# URL da API para o Pokémon com ID 2
url = "https://pokeapi.co/api/v2/pokemon/2/"

# Fazer a solicitação GET para a API
response = requests.get(url)

# Verificar se a solicitação foi bem-sucedida
if response.status_code == 200:
    # Converter a resposta JSON em um dicionário Python
    data = response.json()

    # Extrair os campos desejados
    pokemon_id = data['id']
    name = data['name']
    types = [t['type']['name'] for t in data['types']]
    height = data['height']
    weight = data['weight']
    moves = [move['move']['name'] for move in data['moves']]

    # Exibir as informações
    print(f"Id: {pokemon_id}")
    print(f"Name: {name}")
    print(f"Types: {', '.join(types)}")
    print(f"Height: {height}")
    print(f"Weight: {weight}")
    print(f"Moves: {', '.join(moves)}")
else:
    print(f"Failed to retrieve data: {response.status_code}")
```

```python
import requests
import json

# Definir a faixa de IDs dos Pokémon
pokemon_ids = range(1, 1001)

# Nome do arquivo JSON
output_file = 'pokemon_data.json'

# Inicializar uma lista para armazenar os dados dos Pokémon
pokemon_data = []

# Iterar sobre os IDs dos Pokémon
for pokemon_id in pokemon_ids:
    # URL da API para o Pokémon atual
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_id}/"
    
    # Fazer a solicitação GET para a API
    response = requests.get(url)
    
    # Verificar se a solicitação foi bem-sucedida
    if response.status_code == 200:
        # Converter a resposta JSON em um dicionário Python
        data = response.json()
        
        # Extrair os campos desejados
        pokemon_info = {
            'id': data['id'],
            'name': data['name'],
            'types': [t['type']['name'] for t in data['types']],
            'height': data['height'],
            'weight': data['weight'],
            'moves': [move['move']['name'] for move in data['moves']]
        }
        
        # Adicionar as informações extraídas à lista
        pokemon_data.append(pokemon_info)
        
        # Salvar os dados no arquivo JSON incrementalmente
        with open(output_file, 'w') as f:
            json.dump(pokemon_data, f, indent=4)
        
        print(f"Successfully retrieved data for Pokémon ID {pokemon_id}")
    else:
        print(f"Failed to retrieve data for Pokémon ID {pokemon_id}: {response.status_code}")

print(f"Data saved to {output_file}")
```
